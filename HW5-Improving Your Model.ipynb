### Doing Logistic Regression on the new Data with the Barometric Pressure as a feature

```python
log_names = ['AWND', 'FMTM', 'PGTM', 'TAVG', 'TMAX', 'TMIN', 'TSUN', 'WDF2', 'WDF5', 'WSF2', 'WSF5',  
              'WT03', 'WT07', 'WT08', 'WT10', 'WT11', 'WV03', 'sl-pressure']
X = df2[log_names] # features
y = df2.rain # target variables
```

```python
# splitting the data into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.25, random_state=25)
```

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(max_iter=10000) # Upped the number iter to 10000 from 5000 because it said it couldn't converge

# fit the model with data
logreg.fit(X_train,y_train)

# predict for the test data
y_pred = logreg.predict(X_test)
```

```python
# Compute the accuracy for how well you can predict when it will rain
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100, " %")# the fraction of correctly classified samples.
```

### Retrying Logistic Regression without those features that didn't have much effect

```python
log_names2 = ['AWND', 'TAVG', 'TMAX', 'TMIN', 'WSF2', 'WSF5',  
              'WT03',  'WT08', 'WV03', 'sl-pressure']
X = df2[log_names2] # features
y = df2.rain # target variables

# splitting the data into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.25, random_state=25)

# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(max_iter=10000) # Upped the number iter to 10000 from 5000 because it said it couldn't converge

# fit the model with data
logreg.fit(X_train,y_train)

# predict for the test data
y_pred = logreg.predict(X_test)

# Compute the accuracy for how well you can predict when it will rain
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100, " %")# the fraction of correctly classified samples.
```

The accuracy went up to `82.7%` from `82.6%` by dropping some features that had minimal impact. Let's try some different kinds of validation testing.

### Stratified K-fold Validation

```python
skfold = StratifiedKFold(n_splits=3)
model_skfold = LogisticRegression(max_iter=10000)
results_skfold = model_selection.cross_val_score(model_skfold, X, y, cv=skfold)
print("Accuracy: %.2f%%" % (results_skfold.mean()*100.0))
```

The accuracy went up to `83.1%` from `82.7%` with the stratified K-fold validation.

### Repeated Random Test-Train Splits

```python
kfold2 = model_selection.ShuffleSplit(n_splits=50, test_size=0.25, random_state=100) 
# upped the n_splits to 50 for highest accuracy
model_shufflecv = LogisticRegression(max_iter=10000)
results_4 = model_selection.cross_val_score(model_shufflecv, X, y, cv=kfold2)
print("Accuracy: %.2f%% (%.2f%%)" % (results_4.mean()*100.0, results_4.std()*100.0))
```

The accuracy went up to `83.4%` from `83.1%` with the repeated Random Test-Train Splits.

### Doing Logistic Regression on this data with the Barometric Pressure, Dew Point Temperature, Wind Speed, Etc.

```python
log_names3 = ['HourlyAltimeterSetting',
'HourlyDewPointTemperature',
'HourlyDryBulbTemperature',
'HourlyPressureChange', 
'HourlyPressureTendency', 
'HourlyRelativeHumidity', 
'HourlySeaLevelPressure', 
'HourlyStationPressure', 
'HourlyVisibility', 
'HourlyWetBulbTemperature', 
'HourlyWindDirection', 
'HourlyWindGustSpeed', 
'HourlyWindSpeed' ]
X = df3[log_names3] # features
y = df3.rain # target variables

# splitting the data into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.25, random_state=25) # changed the test-size to .28 to get the best accuracy

# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(max_iter=10000) # Upped the number iter to 10000 from 5000 because it said it couldn't converge

# fit the model with data
logreg.fit(X_train,y_train)

# predict for the test data
y_pred = logreg.predict(X_test)

# Compute the accuracy for how well you can predict when it will rain
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100, " %")# the fraction of correctly classified samples.
```

The accuracy went up to `90.5%` from the original `81.8%` from HW4. This is good enough to satisfy the 5% increase in accuracy. Let's see if we can do better.
